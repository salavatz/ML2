3 Метрики в задачи регрессии

```python
import numpy as np
%pylab inline

x = np.linspace(0, 50, 500)
epsilon = np.random.normal(0, 0.2, 500)
y = np.array([0.5 * x[i] + 1 + epsilon[i] for i in range(0, 500)])
pyplot.figure(figsize(10, 5))
pylab.scatter(x, y, s = 1)
pyplot.xlabel('x')
pyplot.ylabel('y')
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from scipy.optimize import minimize

def mse(data):
    return mean_squared_error([data[0] * i + data[1] for i in x], y)
def f(m, x):
    return [i * m.x[0] + m.x[1] for i in x]
mse_ = minimize(mse, [1, 1])
```

```python
pyplot.figure(figsize(20, 10))
pylab.scatter(x, y, s = 1)
pylab.plot(x, f(mse_, x), 'r')
pyplot.title('mean squared error')
```

```python
x_ = np.linspace(0, 50, 75)
epsilon_ = np.random.normal(0, 0.2, 75)
y_ = np.array([-1 + epsilon_[i] for i in range(0, 75)])
x = np.append(x, x_)
y = np.append(y, y_)
pyplot.figure(figsize(20, 10))
pylab.scatter(x, y, s = 1)
```

```python
def mae(data):
    return mean_absolute_error([data[0] * i + data[1] for i in x], y)
mse_ = minimize(mse, [1,1])
mae_ = minimize(mae, [1,1])
```

```python
pyplot.figure(figsize(20, 10))
pylab.scatter(x, y, s = 1)
pylab.plot(x, f(mse_, x), 'r')
pyplot.title('mean squared error')
```

```python
pyplot.figure(figsize(20, 10))
pylab.scatter(x, y, s = 1)
pylab.plot(x, f(mae_, x), 'r')
pyplot.title('mean absolute error')
```

Из полученных графиков можно сделать вывод, что модель, оптимизирующая MAE, более устойчив к выбросам по сравнению с MSE.
